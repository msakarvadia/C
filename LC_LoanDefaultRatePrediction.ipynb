```
x_Naive = pd.concat([x, y], axis=1)
x_Naive = x_Naive.drop(x_Naive[x_Naive.loan_status.eq(1)].sample(frac=.80).index)
x_Naive = x_Naive.drop(x_Naive.sample(frac=.90).index)
y_Naive = x_Naive['loan_status'].copy()
x_Naive.drop(columns=['loan_status'], inplace=True) 
x_Naive.shape
```

```
#need a function to find class_size
def class_sizeFinder(df, Class):
  return(df[Class].size)

#need function to determine mean of each class
#for Class you need to pass in a df indexed into the class in a way that can be indexed into each individual feature
def mean(df, Class):
  return(df[Class].mean())
  

#need function to determine variance of each class
def stand_dev(df, Class):
  return(df[Class].std())

#need probability density function
```

```
#split into train/test set
X_train, X_test, y_train, y_test = train_test_split(x_Naive, y_Naive, test_size=0.33, random_state=42)
```

```
#split all instances with paid vs defaulted loans

x_one = pd.DataFrame()
x_zero = pd.DataFrame()

k = 0
for i in y_train:
  if i == 1:
    x_one = x_one.append(X_train[k:k+1])
  if i ==0:
    x_zero = x_zero.append(X_train[k:k+1])
  k+=1
```

```
prior_zero = len(x_zero)/len(X_train)
prior_one = len(x_one)/len(X_train)
```

```
#verify that the splitting was accurate
print(x_one.shape)
print(x_zero.shape)
print(x_one.shape[0]+x_zero.shape[0])
y_train.size
```

```
#calculate mean/variance for both classes
mean_one = []
mean_zero = []
std_one = []
std_zero = []

for i in x_zero:
  mean_one.append(mean(x_one, i))
  mean_zero.append(mean(x_zero, i))
  std_one.append(stand_dev(x_one, i))
  std_zero.append(stand_dev(x_one,i))
```

```
#probability density function
import math

def pdf(feature, mean, std):
  a = 1/math.sqrt(2*math.pi*std**2 + 0.000000001)*math.exp(-1/2*((feature-mean)/(std**2 + 0.0001))**2)
  return a
```

```
#calculate predictions for class 1 for test set
test_classOnePred = []
for r in range(len(X_test)): #itterates thru rows
  prob = 1
  for i, k in enumerate(x_one):              #itterates thru classes
    prob = prob * pdf(X_test.iloc[r, i],mean_one[i],std_one[i])
  test_classOnePred.append(prob) 


#calculate predition for class 2 for test set
test_classZeroPred = []
for r in range(len(X_test)): #itterates thru rows
  prob = 1
  for i, k in enumerate(x_one):              #itterates thru classes
    prob = prob * pdf(X_test.iloc[r, i],mean_zero[i],std_zero[i])
  test_classZeroPred.append(prob) 
```

```
#Naive Bayes theorm!!! - values are so small so maybe use Variance instead of STD
# p(a|x)= p(x|a)p(a)/ (p(x|a)p(a)+p(x|c)p(c))

for i in range(len(test_classOnePred)):
  test_classOnePred[i] = test_classOnePred[i]*prior_one/(test_classOnePred[i]*prior_one+test_classZeroPred[i]*prior_zero+.000000000001)

for i in range(len(test_classOnePred)):
  test_classZeroPred[i] = test_classZeroPred[i]*prior_zero/(test_classOnePred[i]*prior_one+test_classZeroPred[i]*prior_zero+.000000000001)

#append correct predictions into a final prediction list

#compare predictions vs actual labels, calculate error
```

```
errorSum = 0

for i in range(len(test_classOnePred)):
  if test_classOnePred[i] > 0.5:
    test_classOnePred[i] = 1
    
for i, k in enumerate(y_test):
  if test_classOnePred[i] == y_test.iloc[k]:
    errorSum += 1
    
print(errorSum/ len(test_classOnePred) * 100)
```
